La prueba de rachas arriba-abajo se utliza para verififcar la independiencia de una secuencia de números.

Esta prueba analiza el patrón de incremento y decrementos entre observaciones consecutivas con el fin de detectar posibles dependencias.  

##HIPOTESIS

$$
H_0: \text{La secuencia de datos es independiente}
$$

$$
H_1: \text{La secuencia de datos NO es independiente}
$$



## PROCEDIMIENTO

Sea una muetra de tamaño $n$ de números pseudoaleatorios.

$$
X_1, X_2, \dots, X_n
$$

La cual es organizada en forma matricial, tomando en cuenta que los datos se leen por columnas.
Si la muestra contiene $n$ observaciones, al comparar valores consecutivos se construye una secuencia de signos $S$ de tamaño $n-1$ definida como:

$$
S_i =
\begin{cases}
+  & \text{si } X_i < X_{i+1} \\
- & \text{si } X_i > X_{i+1}
\end{cases}
\quad \text{para } i = 1,2,\dots,n-1.
$$



Una **racha** se define como una sucesión consecutiva de signos iguales.

Sea $C_0$ el número total de rachas observadas en la secuencia.





## PARÁMETROS TEORICOS

Bajo la hipótesis nula de independencia, el número esperado de rachas tiene:

$$
\mu_{C_0} = \frac{2n - 1}{3}
$$

La varianza esta definida como:
$$
\sigma^2_{C_0} = \frac{16n - 29}{90}
$$

La desviacion estandar
$$
\sigma_{C_0} = \sqrt{\frac{16n - 29}{90}}
$$



## Estadístico de Prueba

El estadístico estandarizado está dado por:

$$
Z_0 = \frac{C_0 - \mu_{C_0}}{\sigma_{C_0}}
$$

## Regla de Decisión

Para un nivel de significancia $\alpha$. Si se cumple la desigualdad, NO rechaza $H_0$ y se concluye que la secuencia puede considerarse independiente.

$$
|Z_0| < Z_{\alpha/2}
$$


$$
-Z_{\alpha/2} <Z_0 < Z_{\alpha/2}
$$

En caso contrario, se rechaza $H_0$ y se concluye que la secuencia presenta independencia.



```python
#importamos las librerias

import numpy as np
from scipy.stats import norm
```

Definiremos una función, la cual recibe como entradas los datos organizados en una matriz, lo cual los leera por columnas.

Posteriormente, se encargara de contruir una secuencia de signos $S$ de tamaño $n-1$, se cuenta el número total de rechas $C_0$, calcula los parémtros teoricos y determina, el estadisiticos de prueba y finalmente toma la decisión si la muestra es independiente o no.





```python
def prueba_rachas (datos, alpha=0.05):
  """
  Parémetros:
      Datos: Colocados en forma matricial
      alpha: Nivel de significancia (0.05)

  Retorna:

        n               -> tamaño de muestra
        rachas          -> número total de rachas observadas (C0)
        media           -> media teórica
        desviacion      -> desviación estándar teórica
        Z0              -> estadístico de prueba
        z_critico       -> valor crítico bilateral
        decision        -> conclusión de la prueba

  """

# PASO 1: leera los número por columnas
  muestra = datos.flatten (order="F")       #order="F" indica lectura por columnas
  n= len (muestra)  # tamaño de la muestra

# PASO 2: Construira la secuencia S tomando en cuenta que el de tamaño n-1
  signos = []  #guardara en un a lista

  for i in range (n-1):
    if muestra[i] < muestra [i+1]:
      signos.append("+")      #aumento
    else:
      signos.append("-")      #disminuyo


#PASO 3: Contar el numero de rachas (C0)
    rachas =1   #al menos existe una racha

  for i in range (1, len(signos)):
    #si el signo cambia, comenzara una nueva racha
    if signos[i] != signos[i-1]:
        rachas += 1




#PASO 4: Calcular los parametros teoricos
  #media teorica
  media = (2*n-1)/3

  #varianza teorica
  varianza = (16*n - 29) / 90

  # Desviación estándar
  desviacion = np.sqrt(varianza)


# PASO 5: calcular el estadistico de Z
  Z_0 = (rachas - media)/ desviacion

  #valor critico
  z_critico = norm.ppf(1- alpha/2)


#PASO 6: Decision estadistica

  if abs(Z_0) < z_critico:
    decision = "No se rechaza H0: los datos pueden considerarse independientes."
  else:
    decision = "Se rechaza H0: existe evidencia de dependencia."

# PASO 7: imprimir los resultados
  return { "n":n, "rachas(C0)": rachas, "media teorica":media , "desviacion estandar": desviacion, "Z_0": Z_0, "valor critico": z_critico, "decision": decision}
```

Mandare a llamar a la función para que utilizando los datos y realice todos los paso para verificar si la muestra safisface la pruba de rachas.

```python
resultado = prueba_rachas(datos)

for clave, valor in resultado.items():
    print(clave, ":", valor)
```
