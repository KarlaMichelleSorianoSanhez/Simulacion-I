Considerando que una distrución uniforme por definición es:

$ U_i ∼ U(0,1)$ para $i=0,1,2, …, n$

En otras palabras
$
f(x) = \begin{cases}
\frac{1}{b-a} & a \le 0 \le b\\
0 & otro \quad caso
\end{cases}
$

donde $a=0$ y $b=1$

Además recordando que si el conjunto de números (datos de la tabla) aleatorios, provienen de una distribución uniforme entonces se debe cumplir:


  - Media

  $$\mu = E[x]= \frac{1}{2}$$

  - Varianza

  $$Var(X)= \mu^2=  \frac{1}{12}$$

  $$ \sigma = \sqrt{\frac{1}{12}} $$


- Para una muestra de tamaño $n$, la **media muestral** o tambien conocida como promedio es:

$$\bar{X}= \frac{1}{n} \sum_{i=1}^{n} X_i $$

$$ \bar{X} = \frac{U_1 + U_2+ \dots U_n}{n} = 0.4826878$$


Dado que la tabla es de *20 filas x 5 columnas*, entonces la muestra es de tamaño $n= 100 $.
El estadistico de prueba es:


$$Z= \frac{\bar{X} - \mu}{\sigma/ \sqrt{n}} $$

o tambien lo podemos escribir de esta manera

$Z_0=\frac{(\bar{X}-0.5)\sqrt{n}}{\sqrt{\frac{1}{12}}}$

$Z_0=\frac{(0.4826878-0.5)10}{\sqrt{\frac{1}{12}}}= -0.599712$


donde n es el número total de datos.



Por otro lado se considera el nivel de confianza $\alpha$, un valor común es $\alpha=0.05$, de modo que puntos criticos se calculan como:

$$
Z_{\alpha/2} = Z_{0.025} =\pm  1.96
$$



Regla de decisión:

* Rechazar $H_0$ si $|Z| > 1.96$

* No rechazar $H_0$ si $|Z| \le 1.96$


### Planteamiento de hipótesis

$$H_0 : \mu= 0.5$$

$$ H_1 : \mu \neq 0.5 $$

Definimos la función para implemetar la prueba de los promedios

```python
def prueba_promedios(data, alpha):

  #numero total de datos
  N= len(data)

  # media muestral
  media= np.mean(data)

  # Desviación estandar para U(0,1)
  sigma = np.sqrt(1/12)

  # Estadistico de prueba
  Z = (media-0.5) * np.sqrt(N)/ sigma

  #valores criticos para la prueba

  z_c = st.norm.ppf(1- alpha/2)

  print(f"Número de datos: {N}")
  print(f"Media muestral:", round(media,6))
  print("Estadístico Z0:", round(Z,6))
  print("Valor crítico ±:", round(z_c,6))

  #decisiones a considerar

  if abs(Z)> z_c:
    print("Decisión: Se rechaza H0.")
    print("Conlusión: La media es diferente de 0.5 y además la muestra no proviene de una distribución U(0,1)")

  else:
     print("Decisión: NO se rechaza H0.")
     print("Conlusión: La media no es diferente de 0.5, por lo que la muestra proviene de una distribución U(0,1)")
```

Se observa que los valores de manera aleatoria alrededor de 0.5. Además la media muestral se escuentra muy cercana a la media teórica sperada para una distribución Uniforme(0,1), lo cual cual es consistente con la pueba estadisitica.



```python
#llamare a la función

alpha= 0.05
prueba_promedios(data, alpha)
```


Lo cual es evidente que $Z_0$ esta dentro del intervalo $[-1.96, 1.96]$ eso quiere decir que está en la región de aceptación de $H_0$ por lo que la decisión es NO rechazar la hipotesis nula.

Eso indica que el conjunto de números provienen de una distrivución uniforme, asi que su media es 0.5. Finalmente pasa la prueba de los Promedios
