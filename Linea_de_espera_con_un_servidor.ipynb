# Simulación de una Línea de Espera con un Servidor (M/M/1)

## Instrucciones

Realizar una simulación de una línea de espera con un servidor utilizando el método de eventos discretos descrito por Sheldon Ross.

Considere un sistema de colas M/M/1 con llegadas gobernadas por un proceso de Poisson de tasa $\lambda$ y tiempos de servicio exponenciales con parámetro $\mu$.

Utilizar como caso de prueba:

$$
\lambda = 4
$$

$$
\mu = 6
$$

El programa debe determinar:

1. El tiempo promedio que un cliente permanece en el sistema.

2. El tiempo en que abandona el sistema el último cliente después del horizonte de observación.

Posteriormente, comparar los resultados obtenidos mediante simulación con los valores teóricos del modelo M/M/1.

## Introducción

Los sistemas de líneas de espera aparecen en una gran variedad de aplicaciones reales, tales como bancos, supermercados, centros de atención telefónica, redes de comunicación y sistemas computacionales.

Uno de los modelos más importantes en teoría de colas es el sistema M/M/1, el cual supone que las llegadas de clientes siguen un proceso de Poisson y que los tiempos de servicio tienen distribución exponencial. Además, existe un único servidor encargado de atender a todos los clientes que llegan al sistema.

En esta práctica se implementará una simulación de eventos discretos siguiendo el procedimiento descrito por Sheldon Ross. La simulación permitirá reproducir el comportamiento dinámico de la cola registrando los tiempos de llegada y salida de los clientes, así como el número de clientes presentes en el sistema en cada instante.

Finalmente, los resultados obtenidos mediante simulación serán comparados con los valores teóricos del modelo M/M/1 para verificar la validez del procedimiento implementado.

## Objetivo

Implementar una simulación de eventos discretos para un sistema de colas M/M/1, con el fin de estimar el tiempo promedio que los clientes permanecen en el sistema y el tiempo de salida del último cliente, comparando posteriormente dichos resultados con las expresiones analíticas de la teoría de colas.

# Solución Analítica

Antes de realizar la simulación, calcularemos los principales indicadores teóricos del sistema de línea de espera con un servidor.

Como las llegadas siguen un proceso de Poisson con tasa $\lambda$ y los tiempos de servicio siguen una distribución exponencial con parámetro $\mu$, el sistema corresponde a un modelo M/M/1.

En este caso:

$$
\lambda = 4
$$

$$
\mu = 6
$$

Para que el sistema sea estable, debe cumplirse:

$$
\lambda < \mu.
$$

Como:

$$
4 < 6,
$$

el sistema es estable. Por lo tanto, se pueden aplicar las fórmulas del modelo M/M/1.

## 1. Factor de utilización del servidor

El factor de utilización representa la proporción de tiempo en la que el servidor permanece ocupado.

Se calcula como:

$$
\rho = \frac{\lambda}{\mu}.
$$

Sustituyendo:

$$
\rho = \frac{4}{6}.
$$

Simplificando:

$$
\rho = \frac{2}{3}.
$$

Por lo tanto:

$$
\rho \approx 0.6667.
$$

Esto significa que el servidor permanece ocupado aproximadamente el $66.67\%$ del tiempo.

## 2. Probabilidad de que no haya clientes en el sistema

La probabilidad de que el sistema esté vacío se denota por $P_0$.

Para un sistema M/M/1:

$$
P_0 = 1-\rho.
$$

Sustituyendo:

$$
P_0 = 1-\frac{2}{3}.
$$

Entonces:

$$
P_0 = \frac{1}{3}.
$$

Por lo tanto:

$$
P_0 \approx 0.3333.
$$

Esto significa que el sistema se encuentra vacío aproximadamente el $33.33\%$ del tiempo.

## 3. Probabilidad de que haya \(n\) clientes en el sistema

Para un sistema M/M/1, la probabilidad de que existan $n$ clientes en el sistema está dada por:

$$
P_n = (1-\rho)\rho^n.
$$

Como:

$$
1-\rho = P_0 = \frac{1}{3}
$$

y

$$
\rho = \frac{2}{3},
$$

entonces:

$$
P_n = \frac{1}{3}\left(\frac{2}{3}\right)^n.
$$

Por ejemplo, para $n=1$:

$$
P_1 = \frac{1}{3}\left(\frac{2}{3}\right)^1.
$$

$$
P_1 = \frac{2}{9}.
$$

$$
P_1 \approx 0.2222.
$$

Para $n=2$:

$$
P_2 = \frac{1}{3}\left(\frac{2}{3}\right)^2.
$$

$$
P_2 = \frac{4}{27}.
$$

$$
P_2 \approx 0.1481.
$$

Para $n=3$:

$$
P_3 = \frac{1}{3}\left(\frac{2}{3}\right)^3.
$$

$$
P_3 = \frac{8}{81}.
$$

$$
P_3 \approx 0.0988.
$$

## 4. Número promedio de clientes en cola

El número promedio de clientes en cola se denota por $L_q$.

Para un sistema M/M/1:

$$
L_q = \frac{\lambda^2}{\mu(\mu-\lambda)}.
$$

Sustituyendo:

$$
L_q = \frac{4^2}{6(6-4)}.
$$

Calculando:

$$
L_q = \frac{16}{6(2)}.
$$

$$
L_q = \frac{16}{12}.
$$

Simplificando:

$$
L_q = \frac{4}{3}.
$$

Por lo tanto:

$$
L_q \approx 1.3333.
$$

Esto significa que, en promedio, hay aproximadamente $1.3333$ clientes esperando en cola.

## 5. Número promedio de clientes en el sistema

El número promedio de clientes en el sistema se denota por $L$.

Para un sistema M/M/1:

$$
L = \frac{\lambda}{\mu-\lambda}.
$$

Sustituyendo:

$$
L = \frac{4}{6-4}.
$$

Calculando:

$$
L = \frac{4}{2}.
$$

Por lo tanto:

$$
L = 2.
$$

Esto significa que, en promedio, hay $2$ clientes dentro del sistema, considerando tanto los que esperan en cola como el cliente que está siendo atendido.

También se puede verificar usando:

$$
L = L_q + \rho.
$$

Sustituyendo:

$$
L = \frac{4}{3} + \frac{2}{3}.
$$

$$
L = \frac{6}{3}.
$$

$$
L = 2.
$$

## 6. Tiempo promedio de espera en cola

El tiempo promedio que un cliente espera en cola se denota por $W_q$.

Por la ley de Little:

$$
W_q = \frac{L_q}{\lambda}.
$$

Sustituyendo:

$$
W_q = \frac{\frac{4}{3}}{4}.
$$

Esto es:

$$
W_q = \frac{4}{3}\cdot \frac{1}{4}.
$$

Entonces:

$$
W_q = \frac{1}{3}.
$$

Por lo tanto:

$$
W_q \approx 0.3333.
$$

Esto significa que, en promedio, un cliente espera aproximadamente $0.3333$ unidades de tiempo antes de iniciar su servicio.

## 7. Tiempo promedio de permanencia en el sistema

El tiempo promedio que un cliente permanece dentro del sistema se denota por $W$.

Este tiempo incluye tanto el tiempo de espera en cola como el tiempo de servicio.

Para un sistema M/M/1:

$$
W = \frac{1}{\mu-\lambda}.
$$

Sustituyendo:

$$
W = \frac{1}{6-4}.
$$

Calculando:

$$
W = \frac{1}{2}.
$$

Por lo tanto:

$$
W = 0.5.
$$

Esto significa que, en promedio, un cliente permanece $0.5$ unidades de tiempo dentro del sistema.

También se puede verificar mediante la ley de Little:

$$
W = \frac{L}{\lambda}.
$$

Sustituyendo:

$$
W = \frac{2}{4}.
$$

$$
W = \frac{1}{2}.
$$

$$
W = 0.5.
$$

# Simulación por Eventos Discretos

Una vez obtenidos los resultados teóricos del modelo M/M/1, se procede a implementar la simulación siguiendo el pseudocódigo propuesto por Sheldon Ross para un sistema de línea de espera con un servidor.

La simulación se basa en el método de eventos discretos. En este enfoque, el reloj de simulación no avanza en intervalos constantes de tiempo, sino que se mueve directamente al instante en que ocurre el siguiente evento relevante para el sistema.

Los eventos considerados son:

1. Llegada de un cliente.
2. Salida de un cliente.

Para describir el estado del sistema se utilizarán las siguientes variables:

$$
t=\text{tiempo actual de simulación},
$$

$$
N_A=\text{número de llegadas hasta el tiempo } t,
$$

$$
N_D=\text{número de salidas hasta el tiempo } t,
$$

$$
n=\text{número de clientes presentes en el sistema}.
$$

La lista de eventos está formada por:

$$
LE=\{t_A,t_D\},
$$

donde:

$$
t_A=\text{tiempo de la próxima llegada},
$$

$$
t_D=\text{tiempo de la próxima salida}.
$$

En cada iteración se compara el valor de \(t_A\) con el valor de \(t_D\). El menor de ellos determina cuál será el siguiente evento que ocurrirá en el sistema.

Además, para cada cliente se registrarán:

$$
A(i)=\text{tiempo de llegada del cliente } i,
$$

$$
D(i)=\text{tiempo de salida del cliente } i,
$$

y

$$
D(i)-A(i),
$$

que representa el tiempo total que el cliente permanece dentro del sistema.

Finalmente, también se calculará:

$$
T_p,
$$

que corresponde al tiempo en que abandona el sistema el último cliente después de haber alcanzado el horizonte de simulación.

## Librerías necesarias

Para implementar la simulación se utilizarán tres librerías.

La librería `random` permitirá generar números aleatorios uniformes en el intervalo \((0,1)\), los cuales serán utilizados para generar los tiempos entre llegadas y los tiempos de servicio.

La librería `math` proporcionará funciones matemáticas necesarias para implementar la transformación inversa de la distribución exponencial.

Finalmente, la librería `tabulate` se utilizará para presentar los resultados en forma de tablas organizadas y fáciles de interpretar.

Estas herramientas son suficientes para implementar completamente el pseudocódigo de Sheldon Ross sin necesidad de utilizar bibliotecas adicionales para análisis de datos.

```python
# ==========================================================
# Librerías necesarias
# ==========================================================

# Generación de números aleatorios
import random as r

# Funciones matemáticas
import math

# Presentación de tablas
from tabulate import tabulate
```

## Generación de tiempos exponenciales

Las llegadas del sistema siguen un proceso de Poisson con tasa \(\lambda\).

Una propiedad importante de los procesos de Poisson es que los tiempos entre llegadas siguen una distribución exponencial.

De manera similar, los tiempos de servicio también se supondrán exponenciales con parámetro \(\mu\).

Para generar una variable aleatoria exponencial se utiliza el método de la transformada inversa.

Si:

$$
U\sim U(0,1),
$$

entonces:

$$
X=-\frac{\ln(U)}{\alpha}
$$

tiene distribución exponencial con parámetro \(\alpha\).

Esta expresión se utilizará tanto para generar tiempos entre llegadas como tiempos de servicio, cambiando únicamente el valor del parámetro correspondiente.

```python
# ==========================================================
# Función para generar variables exponenciales
# ==========================================================

def exponencial(parametro):
    """
    Genera una variable aleatoria exponencial
    usando el método de la transformada inversa.

    Parámetro:
        parametro : tasa de la distribución exponencial

    Regresa:
        Valor aleatorio con distribución exponencial.
    """

    # Generar un número uniforme en (0,1)
    U = r.random()

    # Aplicar la transformada inversa
    X = -math.log(U) / parametro

    # Regresar el valor generado
    return X
```

## Inicialización del algoritmo

Siguiendo el pseudocódigo propuesto por Sheldon Ross, la simulación inicia con el sistema vacío.

Por lo tanto, al tiempo inicial se tiene:

$$
t = 0,
$$

$$
N_A = 0,
$$

$$
N_D = 0,
$$

$$
n = 0.
$$

Además, se genera el tiempo de la primera llegada:

$$
t_A=T_0.
$$

Como inicialmente no existe ningún cliente en servicio, no hay una salida programada. Por esta razón se define:

$$
t_D=\infty.
$$

También se inicializan las estructuras que almacenarán la información de la simulación.

En particular se registrarán:

$$
A(i),
$$

que representa el tiempo de llegada del cliente \(i\),

$$
D(i),
$$

que representa el tiempo de salida del cliente \(i\),

y

$$
D(i)-A(i),
$$

que corresponde al tiempo total que el cliente permanece dentro del sistema.

Asimismo, se almacenará una tabla de eventos para registrar la evolución completa de la simulación.

## Caso 1: ocurre una llegada

El primer caso ocurre cuando la próxima llegada sucede antes que la próxima salida y además ocurre antes del horizonte de simulación.

Matemáticamente:

$$
t_A \leq t_D
$$

y

$$
t_A \leq T.
$$

Cuando esto sucede:

1. El reloj avanza al instante de llegada.

2. Aumenta el número de llegadas:

$$
N_A=N_A+1.
$$

3. Aumenta el número de clientes en el sistema:

$$
n=n+1.
$$

4. Se registra el tiempo de llegada del cliente.

5. Se genera la siguiente llegada.

Si después de la llegada se tiene:

$$
n=1,
$$

significa que el sistema estaba vacío antes de la llegada.

En consecuencia, el cliente entra inmediatamente a servicio y se programa una salida futura.

## Caso 2: ocurre una salida

El segundo caso ocurre cuando la próxima salida sucede antes que la próxima llegada y además ocurre antes del horizonte de simulación.

Matemáticamente:

$$
t_D < t_A
$$

y

$$
t_D \leq T.
$$

Cuando esto sucede:

1. El reloj avanza al instante de salida.

2. Disminuye el número de clientes en el sistema:

$$
n=n-1.
$$

3. Aumenta el número de salidas:

$$
N_D=N_D+1.
$$

4. Se registra el tiempo de salida del cliente.

5. Si después de la salida aún quedan clientes en el sistema, se genera un nuevo tiempo de servicio.

6. Si ya no quedan clientes en el sistema, entonces no hay próxima salida programada y se toma:

$$
t_D=\infty.
$$

## Caso 3: se supera \(T\), pero aún quedan clientes

El tercer caso ocurre cuando el siguiente evento supera el horizonte de simulación \(T\), pero todavía existen clientes dentro del sistema.

Es decir:

$$
\min(t_A,t_D)>T
$$

y

$$
n>0.
$$

En este caso, ya no se aceptan nuevas llegadas, pero el servidor continúa atendiendo a los clientes que quedaron dentro del sistema.

Por lo tanto, se siguen registrando salidas hasta que el sistema quede vacío.

## Caso 4: termina la simulación

El cuarto caso ocurre cuando el siguiente evento supera el horizonte de simulación y el sistema ya se encuentra vacío.

Esto significa que:

$$
\min(t_A,t_D)>T
$$

y

$$
n=0.
$$

En este caso, la simulación termina.

También se calcula:

$$
T_p,
$$

que representa el tiempo en el que sale el último cliente del sistema.

```python
# ==========================================================
# Función principal de simulación
# Línea de espera M/M/1 mediante eventos discretos
# ==========================================================

def simular_ross(lamda, mu, T):

    """
    Simula una línea de espera con un servidor (M/M/1)
    utilizando el algoritmo de eventos discretos.

    Parámetros
    ----------
    lamda : tasa de llegadas

    mu : tasa de servicio

    T : horizonte de observación

    Regresa
    --------
    A       : tiempos de llegada
    D       : tiempos de salida
    eventos : tabla de eventos
    Tp      : tiempo de salida del último cliente
    """

    # ======================================================
    # INICIALIZACIÓN
    # ======================================================

    # Tiempo actual
    t = 0.0

    # Número de llegadas
    NA = 0

    # Número de salidas
    ND = 0

    # Número de clientes en el sistema
    n = 0

    # Primera llegada
    tA = exponencial(lamda)

    # Inicialmente no existe salida programada
    tD = math.inf

    # ======================================================
    # ESTRUCTURAS DE DATOS
    # ======================================================

    # Tiempos de llegada
    A = []

    # Tiempos de salida
    D = []

    # Cola FIFO
    cola = []

    # Registro de eventos
    eventos = []

    # Tiempo de salida del último cliente
    Tp = 0

    # ======================================================
    # CICLO PRINCIPAL
    # ======================================================

    while True:

        # ==================================================
        # CASO 1
        # Ocurre una llegada
        # tA <= tD y tA <= T
        # ==================================================

        if tA <= tD and tA <= T:

            # Avanzar reloj
            t = tA

            # Actualizar número de llegadas
            NA += 1

            # Actualizar clientes en sistema
            n += 1

            # Registrar llegada
            A.append(t)

            # Cliente entra a la cola
            cola.append(NA)

            # Generar siguiente llegada
            tA = t + exponencial(lamda)

            # Si el sistema estaba vacío
            if n == 1:

                # Generar tiempo de servicio
                Y = exponencial(mu)

                # Programar salida
                tD = t + Y

            # Guardar evento
            eventos.append([
                round(t,4),
                "Llegada",
                NA,
                ND,
                n,
                round(tA,4),
                round(tD,4) if tD != math.inf else "inf"
            ])

        # ==================================================
        # CASO 2
        # Ocurre una salida
        # tD < tA y tD <= T
        # ==================================================

        elif tD < tA and tD <= T:

            # Avanzar reloj
            t = tD

            # Sale un cliente
            n -= 1

            # Actualizar salidas
            ND += 1

            # Sale el primero de la cola
            cola.pop(0)

            # Registrar salida
            D.append(t)

            # Si ya no quedan clientes
            if n == 0:

                tD = math.inf

            # Si quedan clientes esperando
            else:

                # Generar siguiente servicio
                Y = exponencial(mu)

                # Programar nueva salida
                tD = t + Y

            # Guardar evento
            eventos.append([
                round(t,4),
                "Salida",
                NA,
                ND,
                n,
                round(tA,4),
                round(tD,4) if tD != math.inf else "inf"
            ])

        # ==================================================
        # CASO 3
        # Se superó T pero quedan clientes
        # min(tA,tD) > T y n > 0
        # ==================================================

        elif min(tA, tD) > T and n > 0:

            # Avanzar reloj
            t = tD

            # Sale un cliente
            n -= 1

            # Actualizar salidas
            ND += 1

            # Eliminar cliente de la cola
            cola.pop(0)

            # Registrar salida
            D.append(t)

            # Si aún quedan clientes
            if n > 0:

                # Generar nuevo servicio
                Y = exponencial(mu)

                # Programar siguiente salida
                tD = t + Y

            # Si el sistema quedó vacío
            else:

                tD = math.inf

            # Guardar evento
            eventos.append([
                round(t,4),
                "Salida posterior a T",
                NA,
                ND,
                n,
                round(tA,4),
                round(tD,4) if tD != math.inf else "inf"
            ])

        # ==================================================
        # CASO 4
        # Se superó T y el sistema quedó vacío
        # min(tA,tD) > T y n = 0
        # ==================================================

        elif min(tA, tD) > T and n == 0:

            # Registrar tiempo de salida
            # del último cliente
            Tp = t

            # Finalizar simulación
            break

    # ======================================================
    # REGRESAR RESULTADOS
    # ======================================================

    return A, D, eventos, Tp
```

## Ejecución de la simulación

Una vez implementado el algoritmo de eventos discretos, se procede a ejecutar la simulación utilizando los parámetros especificados en el problema.

Se consideran:

$$
\lambda = 4
$$

y

$$
\mu = 6.
$$

Además, se fija un horizonte de observación \(T\). Durante este intervalo se aceptan nuevas llegadas al sistema. Una vez alcanzado dicho horizonte, el algoritmo continúa procesando las salidas pendientes hasta que todos los clientes abandonen el sistema.

Con el fin de obtener resultados reproducibles, se fija una semilla para el generador de números aleatorios.

```python
# ==========================================================
# Parámetros del problema
# ==========================================================

lamda = 4
mu = 6

# Horizonte de observación
T = 100

# Semilla para reproducibilidad
r.seed(123)

# ==========================================================
# Ejecutar simulación
# ==========================================================

A, D, eventos, Tp = simular_ross(lamda, mu, T)

print("Simulación terminada correctamente.")
```

## Tabla de clientes

Siguiendo la notación utilizada por Sheldon Ross, para cada cliente se registra:

$$
A(i)
$$

Tiempo de llegada del cliente \(i\).

$$
D(i)
$$

Tiempo de salida del cliente \(i\).

Asimismo, se calcula:

$$
D(i)-A(i),
$$

que representa el tiempo total que el cliente permaneció dentro del sistema.

Esta información permitirá calcular el tiempo promedio de permanencia en el sistema y compararlo con el valor teórico obtenido anteriormente.

```python
# ==========================================================
# Construcción de la tabla de clientes
# ==========================================================

clientes = []

for i in range(len(D)):

    tiempo_sistema = D[i] - A[i]

    clientes.append([
        i + 1,
        round(A[i], 4),
        round(D[i], 4),
        round(tiempo_sistema, 4)
    ])

print(
    tabulate(
        clientes[:20],
        headers=[
            "Cliente",
            "A(i)",
            "D(i)",
            "D(i)-A(i)"
        ],
        tablefmt="fancy_grid"
    )
)
```
