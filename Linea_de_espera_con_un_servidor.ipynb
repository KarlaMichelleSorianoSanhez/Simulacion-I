Uno de los modelos mГЎs importantes en teorГӯa de colas es el sistema M/M/1, el cual supone que las llegadas de clientes siguen un proceso de Poisson y que los tiempos de servicio tienen distribuciГіn exponencial. Finalmente, los resultados obtenidos mediante simulaciГіn serГЎn comparados con los valores teГіricos del modelo M/M/1 para verificar la validez del procedimiento implementado.

## Objetivo

Implementar una simulaciГіn de eventos discretos para un sistema de colas M/M/1, con el fin de estimar el tiempo promedio que los clientes permanecen en el sistema y el tiempo de salida del Гәltimo cliente, comparando posteriormente dichos resultados con las expresiones analГӯticas de la teorГӯa de colas.

# SoluciГіn AnalГӯtica

Antes de realizar la simulaciГіn, calcularemos los principales indicadores teГіricos del sistema de lГӯnea de espera con un servidor.

Como las llegadas siguen un proceso de Poisson con tasa $\lambda$ y los tiempos de servicio siguen una distribuciГіn exponencial con parГЎmetro $\mu$, el sistema corresponde a un modelo M/M/1.

En este caso:

$$
\lambda = 4
$$

$$
\mu = 6
$$

Para que el sistema sea estable, debe cumplirse:

$$
\lambda < \mu.
$$

Como:

$$
4 < 6,
$$

el sistema es estable. Por lo tanto, se pueden aplicar las fГіrmulas del modelo M/M/1.

## 1. Factor de utilizaciГіn del servidor

El factor de utilizaciГіn representa la proporciГіn de tiempo en la que el servidor permanece ocupado.

Se calcula como:

$$
\rho = \frac{\lambda}{\mu}.
$$

Sustituyendo:

$$
\rho = \frac{4}{6}.
$$

Simplificando:

$$
\rho = \frac{2}{3}.
$$

Por lo tanto:

$$
\rho \approx 0.6667.
$$

Esto significa que el servidor permanece ocupado aproximadamente el $66.67\%$ del tiempo.

## 2. Probabilidad de que no haya clientes en el sistema

La probabilidad de que el sistema estГ© vacГӯo se denota por $P_0$.

Para un sistema M/M/1:

$$
P_0 = 1-\rho.
$$

Sustituyendo:

$$
P_0 = 1-\frac{2}{3}.
$$

Entonces:

$$
P_0 = \frac{1}{3}.
$$

Por lo tanto:

$$
P_0 \approx 0.3333.
$$

Esto significa que el sistema se encuentra vacГӯo aproximadamente el $33.33\%$ del tiempo.

## 3. Probabilidad de que haya \(n\) clientes en el sistema

Para un sistema M/M/1, la probabilidad de que existan $n$ clientes en el sistema estГЎ dada por:

$$
P_n = (1-\rho)\rho^n.
$$

Como:

$$
1-\rho = P_0 = \frac{1}{3}
$$

y

$$
\rho = \frac{2}{3},
$$

entonces:

$$
P_n = \frac{1}{3}\left(\frac{2}{3}\right)^n.
$$

Por ejemplo, para $n=1$:

$$
P_1 = \frac{1}{3}\left(\frac{2}{3}\right)^1.
$$

$$
P_1 = \frac{2}{9}.
$$

$$
P_1 \approx 0.2222.
$$

Para $n=2$:

$$
P_2 = \frac{1}{3}\left(\frac{2}{3}\right)^2.
$$

$$
P_2 = \frac{4}{27}.
$$

$$
P_2 \approx 0.1481.
$$

Para $n=3$:

$$
P_3 = \frac{1}{3}\left(\frac{2}{3}\right)^3.
$$

$$
P_3 = \frac{8}{81}.
$$

$$
P_3 \approx 0.0988.
$$

## 4. NГәmero promedio de clientes en cola

El nГәmero promedio de clientes en cola se denota por $L_q$.

Para un sistema M/M/1:

$$
L_q = \frac{\lambda^2}{\mu(\mu-\lambda)}.
$$

Sustituyendo:

$$
L_q = \frac{4^2}{6(6-4)}.
$$

Calculando:

$$
L_q = \frac{16}{6(2)}.
$$

$$
L_q = \frac{16}{12}.
$$

Simplificando:

$$
L_q = \frac{4}{3}.
$$

Por lo tanto:

$$
L_q \approx 1.3333.
$$

Esto significa que, en promedio, hay aproximadamente $1.3333$ clientes esperando en cola.

## 5. NГәmero promedio de clientes en el sistema

El nГәmero promedio de clientes en el sistema se denota por $L$.

Para un sistema M/M/1:

$$
L = \frac{\lambda}{\mu-\lambda}.
$$

Sustituyendo:

$$
L = \frac{4}{6-4}.
$$

Calculando:

$$
L = \frac{4}{2}.
$$

Por lo tanto:

$$
L = 2.
$$

Esto significa que, en promedio, hay $2$ clientes dentro del sistema, considerando tanto los que esperan en cola como el cliente que estГЎ siendo atendido.

TambiГ©n se puede verificar usando:

$$
L = L_q + \rho.
$$

Sustituyendo:

$$
L = \frac{4}{3} + \frac{2}{3}.
$$

$$
L = \frac{6}{3}.
$$

$$
L = 2.
$$

## 6. Tiempo promedio de espera en cola

El tiempo promedio que un cliente espera en cola se denota por $W_q$.

Por la ley de Little:

$$
W_q = \frac{L_q}{\lambda}.
$$

Sustituyendo:

$$
W_q = \frac{\frac{4}{3}}{4}.
$$

Esto es:

$$
W_q = \frac{4}{3}\cdot \frac{1}{4}.
$$

Entonces:

$$
W_q = \frac{1}{3}.
$$

Por lo tanto:

$$
W_q \approx 0.3333.
$$

Esto significa que, en promedio, un cliente espera aproximadamente $0.3333$ unidades de tiempo antes de iniciar su servicio.

## 7. Tiempo promedio de permanencia en el sistema

El tiempo promedio que un cliente permanece dentro del sistema se denota por $W$.

Este tiempo incluye tanto el tiempo de espera en cola como el tiempo de servicio.

Para un sistema M/M/1:

$$
W = \frac{1}{\mu-\lambda}.
$$

Sustituyendo:

$$
W = \frac{1}{6-4}.
$$

Calculando:

$$
W = \frac{1}{2}.
$$

Por lo tanto:

$$
W = 0.5.
$$

Esto significa que, en promedio, un cliente permanece $0.5$ unidades de tiempo dentro del sistema.

TambiГ©n se puede verificar mediante la ley de Little:

$$
W = \frac{L}{\lambda}.
$$

Sustituyendo:

$$
W = \frac{2}{4}.
$$

$$
W = \frac{1}{2}.
$$

$$
W = 0.5.
$$

# SimulaciГіn por Eventos Discretos

Una vez obtenidos los resultados teГіricos del modelo M/M/1, se procede a implementar la simulaciГіn siguiendo el pseudocГіdigo propuesto por Sheldon Ross para un sistema de lГӯnea de espera con un servidor.

La simulaciГіn se basa en el mГ©todo de eventos discretos. En este enfoque, el reloj de simulaciГіn no avanza en intervalos constantes de tiempo, sino que se mueve directamente al instante en que ocurre el siguiente evento relevante para el sistema.

Los eventos considerados son:

1. Llegada de un cliente.
2. Salida de un cliente.

Para describir el estado del sistema se utilizarГЎn las siguientes variables:

$$
t=\text{tiempo actual de simulaciГіn},
$$

$$
N_A=\text{nГәmero de llegadas hasta el tiempo } t,
$$

$$
N_D=\text{nГәmero de salidas hasta el tiempo } t,
$$

$$
n=\text{nГәmero de clientes presentes en el sistema}.
$$

La lista de eventos estГЎ formada por:

$$
LE=\{t_A,t_D\},
$$

donde:

$$
t_A=\text{tiempo de la prГіxima llegada},
$$

$$
t_D=\text{tiempo de la prГіxima salida}.
$$

En cada iteraciГіn se compara el valor de \(t_A\) con el valor de \(t_D\). El menor de ellos determina cuГЎl serГЎ el siguiente evento que ocurrirГЎ en el sistema.

AdemГЎs, para cada cliente se registrarГЎn:

$$
A(i)=\text{tiempo de llegada del cliente } i,
$$

$$
D(i)=\text{tiempo de salida del cliente } i,
$$

y

$$
D(i)-A(i),
$$

que representa el tiempo total que el cliente permanece dentro del sistema.

Finalmente, tambiГ©n se calcularГЎ:

$$
T_p,
$$

que corresponde al tiempo en que abandona el sistema el Гәltimo cliente despuГ©s de haber alcanzado el horizonte de simulaciГіn.

## LibrerГӯas necesarias

Para implementar la simulaciГіn se utilizarГЎn tres librerГӯas.

La librerГӯa `random` permitirГЎ generar nГәmeros aleatorios uniformes en el intervalo \((0,1)\), los cuales serГЎn utilizados para generar los tiempos entre llegadas y los tiempos de servicio.

La librerГӯa `math` proporcionarГЎ funciones matemГЎticas necesarias para implementar la transformaciГіn inversa de la distribuciГіn exponencial.

Finalmente, la librerГӯa `tabulate` se utilizarГЎ para presentar los resultados en forma de tablas organizadas y fГЎciles de interpretar.

Estas herramientas son suficientes para implementar completamente el pseudocГіdigo de Sheldon Ross sin necesidad de utilizar bibliotecas adicionales para anГЎlisis de datos.

```python
# ==========================================================
# LibrerГӯas necesarias
# ==========================================================

# GeneraciГіn de nГәmeros aleatorios
import random as r

# Funciones matemГЎticas
import math

# PresentaciГіn de tablas
from tabulate import tabulate
```

## GeneraciГіn de tiempos exponenciales

Las llegadas del sistema siguen un proceso de Poisson con tasa \(\lambda\).

Una propiedad importante de los procesos de Poisson es que los tiempos entre llegadas siguen una distribuciГіn exponencial.

De manera similar, los tiempos de servicio tambiГ©n se supondrГЎn exponenciales con parГЎmetro \(\mu\).

Para generar una variable aleatoria exponencial se utiliza el mГ©todo de la transformada inversa.

Si:

$$
U\sim U(0,1),
$$

entonces:

$$
X=-\frac{\ln(U)}{\alpha}
$$

tiene distribuciГіn exponencial con parГЎmetro \(\alpha\).

Esta expresiГіn se utilizarГЎ tanto para generar tiempos entre llegadas como tiempos de servicio, cambiando Гәnicamente el valor del parГЎmetro correspondiente.

```python
# ==========================================================
# FunciГіn para generar variables exponenciales
# ==========================================================

def exponencial(parametro):
    """
    Genera una variable aleatoria exponencial
    usando el mГ©todo de la transformada inversa.

    ParГЎmetro:
        parametro : tasa de la distribuciГіn exponencial

    Regresa:
        Valor aleatorio con distribuciГіn exponencial.
    """

    # Generar un nГәmero uniforme en (0,1)
    U = r.random()

    # Aplicar la transformada inversa
    X = -math.log(U) / parametro

    # Regresar el valor generado
    return X
```

## InicializaciГіn del algoritmo

Siguiendo el pseudocГіdigo propuesto por Sheldon Ross, la simulaciГіn inicia con el sistema vacГӯo.

Por lo tanto, al tiempo inicial se tiene:

$$
t = 0,
$$

$$
N_A = 0,
$$

$$
N_D = 0,
$$

$$
n = 0.
$$

AdemГЎs, se genera el tiempo de la primera llegada:

$$
t_A=T_0.
$$

Como inicialmente no existe ningГәn cliente en servicio, no hay una salida programada. Por esta razГіn se define:

$$
t_D=\infty.
$$

TambiГ©n se inicializan las estructuras que almacenarГЎn la informaciГіn de la simulaciГіn.

En particular se registrarГЎn:

$$
A(i),
$$

que representa el tiempo de llegada del cliente \(i\),

$$
D(i),
$$

que representa el tiempo de salida del cliente \(i\),

y

$$
D(i)-A(i),
$$

que corresponde al tiempo total que el cliente permanece dentro del sistema.

Asimismo, se almacenarГЎ una tabla de eventos para registrar la evoluciГіn completa de la simulaciГіn.

## Caso 1: ocurre una llegada

El primer caso ocurre cuando la prГіxima llegada sucede antes que la prГіxima salida y ademГЎs ocurre antes del horizonte de simulaciГіn.

MatemГЎticamente:

$$
t_A \leq t_D
$$

y

$$
t_A \leq T.
$$

Cuando esto sucede:

1. El reloj avanza al instante de llegada.

2. Aumenta el nГәmero de llegadas:

$$
N_A=N_A+1.
$$

3. Aumenta el nГәmero de clientes en el sistema:

$$
n=n+1.
$$

4. Se registra el tiempo de llegada del cliente.

5. Se genera la siguiente llegada.

Si despuГ©s de la llegada se tiene:

$$
n=1,
$$

significa que el sistema estaba vacГӯo antes de la llegada.

En consecuencia, el cliente entra inmediatamente a servicio y se programa una salida futura.

## Caso 2: ocurre una salida

El segundo caso ocurre cuando la prГіxima salida sucede antes que la prГіxima llegada y ademГЎs ocurre antes del horizonte de simulaciГіn.

MatemГЎticamente:

$$
t_D < t_A
$$

y

$$
t_D \leq T.
$$

Cuando esto sucede:

1. El reloj avanza al instante de salida.

2. Disminuye el nГәmero de clientes en el sistema:

$$
n=n-1.
$$

3. Aumenta el nГәmero de salidas:

$$
N_D=N_D+1.
$$

4. Se registra el tiempo de salida del cliente.

5. Si despuГ©s de la salida aГәn quedan clientes en el sistema, se genera un nuevo tiempo de servicio.

6. Si ya no quedan clientes en el sistema, entonces no hay prГіxima salida programada y se toma:

$$
t_D=\infty.
$$

## Caso 3: se supera \(T\), pero aГәn quedan clientes

El tercer caso ocurre cuando el siguiente evento supera el horizonte de simulaciГіn \(T\), pero todavГӯa existen clientes dentro del sistema.

Es decir:

$$
\min(t_A,t_D)>T
$$

y

$$
n>0.
$$

En este caso, ya no se aceptan nuevas llegadas, pero el servidor continГәa atendiendo a los clientes que quedaron dentro del sistema.

Por lo tanto, se siguen registrando salidas hasta que el sistema quede vacГӯo.

## Caso 4: termina la simulaciГіn

El cuarto caso ocurre cuando el siguiente evento supera el horizonte de simulaciГіn y el sistema ya se encuentra vacГӯo.

Esto significa que:

$$
\min(t_A,t_D)>T
$$

y

$$
n=0.
$$

En este caso, la simulaciГіn termina.

TambiГ©n se calcula:

$$
T_p,
$$

que representa el tiempo en el que sale el Гәltimo cliente del sistema.

```python
# ==========================================================
# FunciГіn principal de simulaciГіn
# LГӯnea de espera M/M/1 mediante eventos discretos
# ==========================================================

def simular_ross(lamda, mu, T):

    """
    Simula una lГӯnea de espera con un servidor (M/M/1)
    utilizando el algoritmo de eventos discretos.

    ParГЎmetros
    ----------
    lamda : tasa de llegadas

    mu : tasa de servicio

    T : horizonte de observaciГіn

    Regresa
    --------
    A       : tiempos de llegada
    D       : tiempos de salida
    eventos : tabla de eventos
    Tp      : tiempo de salida del Гәltimo cliente
    """

    # ======================================================
    # INICIALIZACIГ“N
    # ======================================================

    # Tiempo actual
    t = 0.0

    # NГәmero de llegadas
    NA = 0

    # NГәmero de salidas
    ND = 0

    # NГәmero de clientes en el sistema
    n = 0

    # Primera llegada
    tA = exponencial(lamda)

    # Inicialmente no existe salida programada
    tD = math.inf

    # ======================================================
    # ESTRUCTURAS DE DATOS
    # ======================================================

    # Tiempos de llegada
    A = []

    # Tiempos de salida
    D = []

    # Cola FIFO
    cola = []

    # Registro de eventos
    eventos = []

    # Tiempo de salida del Гәltimo cliente
    Tp = 0

    # ======================================================
    # CICLO PRINCIPAL
    # ======================================================

    while True:

        # ==================================================
        # CASO 1
        # Ocurre una llegada
        # tA <= tD y tA <= T
        # ==================================================

        if tA <= tD and tA <= T:

            # Avanzar reloj
            t = tA

            # Actualizar nГәmero de llegadas
            NA += 1

            # Actualizar clientes en sistema
            n += 1

            # Registrar llegada
            A.append(t)

            # Cliente entra a la cola
            cola.append(NA)

            # Generar siguiente llegada
            tA = t + exponencial(lamda)

            # Si el sistema estaba vacГӯo
            if n == 1:

                # Generar tiempo de servicio
                Y = exponencial(mu)

                # Programar salida
                tD = t + Y

            # Guardar evento
            eventos.append([
                round(t,4),
                "Llegada",
                NA,
                ND,
                n,
                round(tA,4),
                round(tD,4) if tD != math.inf else "inf"
            ])

        # ==================================================
        # CASO 2
        # Ocurre una salida
        # tD < tA y tD <= T
        # ==================================================

        elif tD < tA and tD <= T:

            # Avanzar reloj
            t = tD

            # Sale un cliente
            n -= 1

            # Actualizar salidas
            ND += 1

            # Sale el primero de la cola
            cola.pop(0)

            # Registrar salida
            D.append(t)

            # Si ya no quedan clientes
            if n == 0:

                tD = math.inf

            # Si quedan clientes esperando
            else:

                # Generar siguiente servicio
                Y = exponencial(mu)

                # Programar nueva salida
                tD = t + Y

            # Guardar evento
            eventos.append([
                round(t,4),
                "Salida",
                NA,
                ND,
                n,
                round(tA,4),
                round(tD,4) if tD != math.inf else "inf"
            ])

        # ==================================================
        # CASO 3
        # Se superГі T pero quedan clientes
        # min(tA,tD) > T y n > 0
        # ==================================================

        elif min(tA, tD) > T and n > 0:

            # Avanzar reloj
            t = tD

            # Sale un cliente
            n -= 1

            # Actualizar salidas
            ND += 1

            # Eliminar cliente de la cola
            cola.pop(0)

            # Registrar salida
            D.append(t)

            # Si aГәn quedan clientes
            if n > 0:

                # Generar nuevo servicio
                Y = exponencial(mu)

                # Programar siguiente salida
                tD = t + Y

            # Si el sistema quedГі vacГӯo
            else:

                tD = math.inf

            # Guardar evento
            eventos.append([
                round(t,4),
                "Salida posterior a T",
                NA,
                ND,
                n,
                round(tA,4),
                round(tD,4) if tD != math.inf else "inf"
            ])

        # ==================================================
        # CASO 4
        # Se superГі T y el sistema quedГі vacГӯo
        # min(tA,tD) > T y n = 0
        # ==================================================

        elif min(tA, tD) > T and n == 0:

            # Registrar tiempo de salida
            # del Гәltimo cliente
            Tp = t

            # Finalizar simulaciГіn
            break

    # ======================================================
    # REGRESAR RESULTADOS
    # ======================================================

    return A, D, eventos, Tp
```

## EjecuciГіn de la simulaciГіn

Una vez implementado el algoritmo de eventos discretos, se procede a ejecutar la simulaciГіn utilizando los parГЎmetros especificados en el problema.

Se consideran:

$$
\lambda = 4
$$

y

$$
\mu = 6.
$$

AdemГЎs, se fija un horizonte de observaciГіn \(T\). Durante este intervalo se aceptan nuevas llegadas al sistema. Una vez alcanzado dicho horizonte, el algoritmo continГәa procesando las salidas pendientes hasta que todos los clientes abandonen el sistema.

Con el fin de obtener resultados reproducibles, se fija una semilla para el generador de nГәmeros aleatorios.

```python
# ==========================================================
# ParГЎmetros del problema
# ==========================================================

lamda = 4
mu = 6

# Horizonte de observaciГіn
T = 100

# Semilla para reproducibilidad
r.seed(123)

# ==========================================================
# Ejecutar simulaciГіn
# ==========================================================

A, D, eventos, Tp = simular_ross(lamda, mu, T)

print("SimulaciГіn terminada correctamente.")
```

## Tabla de clientes

Siguiendo la notaciГіn utilizada por Sheldon Ross, para cada cliente se registra:

$$
A(i)
$$

Tiempo de llegada del cliente \(i\).

$$
D(i)
$$

Tiempo de salida del cliente \(i\).

Asimismo, se calcula:

$$
D(i)-A(i),
$$

que representa el tiempo total que el cliente permaneciГі dentro del sistema.

Esta informaciГіn permitirГЎ calcular el tiempo promedio de permanencia en el sistema y compararlo con el valor teГіrico obtenido anteriormente.

```python
# ==========================================================
# ConstrucciГіn de la tabla de clientes
# ==========================================================

clientes = []

for i in range(len(D)):

    tiempo_sistema = D[i] - A[i]

    clientes.append([
        i + 1,
        round(A[i], 4),
        round(D[i], 4),
        round(tiempo_sistema, 4)
    ])

print(
    tabulate(
        clientes[:20],
        headers=[
            "Cliente",
            "A(i)",
            "D(i)",
            "D(i)-A(i)"
        ],
        tablefmt="fancy_grid"
    )
)
```

## Inciso (a): Tiempo promedio de un cliente en el sistema

Una vez obtenidos los tiempos de llegada y salida de todos los clientes, podemos calcular cuГЎnto tiempo permaneciГі cada cliente dentro del sistema.

Para el cliente \(i\), el tiempo total en el sistema estГЎ dado por:

$$
W_i=D(i)-A(i)
$$

donde:

$$
A(i)
$$

es el tiempo de llegada del cliente \(i\),

y

$$
D(i)
$$

es el tiempo de salida del cliente $i$.

Por lo tanto, el tiempo promedio que los clientes permanecen en el sistema se obtiene mediante:

$$
\bar W=
\frac{1}{N_D}
\sum_{i=1}^{N_D}
\left[D(i)-A(i)\right].
$$

Este valor representa el tiempo promedio que un cliente pasa dentro del sistema, considerando tanto el tiempo de espera en la cola como el tiempo de servicio.

```python
# ==========================================================
# Inciso (a)
# Tiempo promedio de un cliente en el sistema
# ==========================================================

# Lista donde guardaremos D(i)-A(i)
tiempos_sistema = []

for i in range(len(D)):

    # Tiempo total que el cliente i permaneciГі en el sistema
    tiempo_cliente = D[i] - A[i]

    tiempos_sistema.append(tiempo_cliente)

# Promedio de todos los tiempos
W_promedio = sum(tiempos_sistema) / len(tiempos_sistema)

print("Tiempo promedio en el sistema:")
print(round(W_promedio, 4))
```

El tiempo promedio obtenido mediante simulaciГіn fue:

$$
\bar W = 0.3768.
$$

Este valor representa el tiempo promedio que los clientes permanecieron dentro del sistema durante la corrida realizada.

Debido a la naturaleza aleatoria del proceso de llegadas y servicios, el resultado obtenido puede diferir del valor teГіrico esperado.

Posteriormente se realizarГЎ una comparaciГіn con el modelo M/M/1 para analizar quГ© tan cercana es la simulaciГіn al comportamiento teГіrico.

## Factor de utilizaciГіn

El factor de utilizaciГіn mide la proporciГіn del tiempo en que el servidor permanece ocupado.

Se calcula mediante:

$$
\rho=\frac{\lambda}{\mu}.
$$

Sustituyendo:

$$
\rho=\frac{4}{6}
=
0.6667.
$$

Esto significa que el servidor permanece ocupado aproximadamente el \(66.67\%\) del tiempo.

```python
# ==========================================================
# Factor de utilizaciГіn
# ==========================================================

rho = lamda / mu

print("rho =", round(rho, 4))
```

## Probabilidad de sistema vacГӯo

La probabilidad de que no existan clientes en el sistema es:

$$
P_0 = 1-\rho.
$$

Por lo tanto:

$$
P_0
=
1-0.6667
=
0.3333.
$$

```python
# ==========================================================
# Probabilidad de sistema vacГӯo
# ==========================================================

P0 = 1 - rho

print("P0 =", round(P0, 4))
```

## NГәmero promedio de clientes en cola

La longitud promedio de la cola estГЎ dada por:

$$
L_q
=
\frac{\lambda^2}
{\mu(\mu-\lambda)}.
$$

Sustituyendo:

$$
L_q
=
\frac{4^2}
{6(6-4)}
=
1.3333.
$$

En promedio existen aproximadamente \(1.33\) clientes esperando servicio.

```python
# ==========================================================
# NГәmero promedio en cola
# ==========================================================

Lq = (lamda**2) / (mu * (mu - lamda))

print("Lq =", round(Lq, 4))
```

## NГәmero promedio de clientes en el sistema

La cantidad promedio de clientes en el sistema es:

$$
L_s
=
L_q
+

L_s
=
1.3333
+
0.6667

2.
$$

```python
# ==========================================================
# NГәmero promedio en el sistema
# ==========================================================

Ls = Lq + rho

print("Ls =", round(Ls, 4))
```

## Tiempo promedio en cola

El tiempo promedio que un cliente espera antes de iniciar su servicio es:

$$
W_q

\frac{L_q}{\lambda}.

W_q

```python
# ==========================================================
# Tiempo promedio en cola
# ==========================================================

Wq = Lq / lamda

print("Wq =", round(Wq, 4))
```

La medida teГіrica correspondiente al inciso (a) es:
